### Matches

```python
matches_url = "https://api.opendota.com/api/matches"
```

```python
matchid = 271145478
matchid_url = matches_url + "/" + str(matchid)
```

```python
rendered_matches = []
for i in range(20):
    try:
        data = urllib.request.urlopen(matchid_url)
        data = data.read().decode("utf-8")
        data = json.loads(data)
        rendered_matches.append(data)
        time.sleep(1)
    except Exception as e:
        print(str(e))
        time.sleep(1)
    matchid+=1
```

```python
df = pd.DataFrame(rendered_matches)
df.head()
```
